```python
import pandas as pd
import numpy as np
```

## Results importing

```python
results = pd.read_csv("results - word vectorization.csv")

results_idf = pd.read_csv("results - word vectorization - IDF.csv")

results_pos = pd.read_csv("results - word vectorization - POS.csv")
```

## Comparison of results by arithmetics	

```python
results_arithmetics = results.drop(columns=["questions modification", "metric"])

results_arithmetics_grouped = results_arithmetics.groupby(["arithmetics"]).mean()

results_arithmetics_grouped = results_arithmetics_grouped.round({"accuracy [%]": 2, "average execution time [s]": 4})
results_arithmetics_grouped["average sentence rank"] = results_arithmetics_grouped["average sentence rank"].astype(int)


results_arithmetics_grouped.head()
```

## Comparison of results by method

```python
results.insert(0, "method",pd.Series(['/' for x in range(len(results.index))]))
results_idf.insert(0, "method", pd.Series(["IDF" for x in range(len(results_idf.index))]))
results_pos.insert(0, "method", pd.Series(["POS" for x in range(len(results_pos.index))]))

results_total = pd.concat([results, results_idf, results_pos], axis=0)

results_method = results_total.drop(columns=["arithmetics", "questions modification", "metric"])

results_method_grouped = results_method.groupby(["method"]).mean()

results_method_grouped = results_method_grouped.round({"accuracy [%]": 2, "average execution time [s]": 4})
results_method_grouped["average sentence rank"] = results_method_grouped["average sentence rank"].astype(int)

results_method_grouped.head()
```
