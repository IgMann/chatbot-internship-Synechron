## Importing libraries 

```python
import pandas as pd
import numpy as np
```

## Results importing

```python
results = pd.read_csv("tf-idf - results.csv")

results_bigrams = pd.read_csv("tf-idf - results - bigrams.csv")
```

```python
results.head(60)
```

## Showing results with bigrams

```python
results_bigrams.head(60)
```

## Comparison of results without bigrams by method

```python
results_method = results.drop(columns=["word simplification", "questions modification", "metric"])

results_method_grouped = results_method.groupby(["method"]).mean()

results_method_grouped = results_method_grouped.round({"average sentence rank": 0, "average execution time [s]": 4})
results_method_grouped["average sentence rank"] = results_method_grouped["average sentence rank"].astype(int)


results_method_grouped.head()
```

## Comparison of results with and without bigrams

```python
results.insert(0, "n-grams",pd.Series(["unigram" for x in range(len(results.index))]))
results_bigrams.insert(0, "n-grams", pd.Series(["bigram" for x in range(len(results_bigrams.index))]))

results_total = pd.concat([results, results_bigrams], axis=0)

results_n_grams = results_total.drop(columns=["method", "word simplification", "questions modification", "metric"])

results_n_grams_grouped = results_n_grams.groupby(["n-grams"]).mean()

results_n_grams_grouped = results_n_grams_grouped.round({"average sentence rank": 0, "average execution time [s]": 4})
results_n_grams_grouped["average sentence rank"] = results_n_grams_grouped["average sentence rank"].astype(int)

results_n_grams_grouped.head()
```
